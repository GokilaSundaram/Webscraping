```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Web page URL#
try:
    url='https://www.studentcrowd.com/best-halls-l1004971-s1008369-nottingham_trent_university-nottingham'

    page=requests.get(url)

    source=BeautifulSoup(page.text,'html')
    print(source)

except Exception as e:
    print(e)
```

```python
# To Find the header tag and respective class#
title= source.find('h2')
accomodationlist= source.find('ul', class_='tw-mt-8 tw-mx-0 tw-mb-0 list-style--none').find_all('li',itemprop='itemListElement')
```

```python
print(accomodationlist)
```

```python
# Create a List for the Titles#
accommodation_data = {'Rank': [], 'Name': [], 'Price': [], 'Rating': []}
```

```python
# Loop through the html content and find the text to be printed#
for accomodation in accomodationlist:
    rank=accomodation.find('div',class_='top-table__row__image-overlay').b.text.strip('#')
    price_from=accomodation.find('div',class_='price-from')
    price=price_from.text
    rating=accomodation.find('div',class_='epsilon c-green mt0 hide sm-show').b.text
    name=accomodation.find('h2',class_='mb0').span.text
    review=accomodation.find('span',class_='top-table__row__quote').text
    accommodation_data['Rank'].append(rank)
    accommodation_data['Name'].append(name)
    accommodation_data['Price'].append(price)
    accommodation_data['Rating'].append(rating)
```

```python
print(accommodation_data)
```

```python
# Create a Dataframe and upload the extracted data#

accommodation_df=pd.DataFrame(accommodation_data)
```

```python
print(accommodation_df)
```

```python
# Save the extracted data as csv file#
file_path=r'/kaggle/working/NTU_AccomodationList.csv'
accommodation_df.to_csv(file_path,index=None)
```
